```
import stanza
stanza.download('ru')
def stanza_nlp_ru(text):
  nlp = stanza.Pipeline(lang='ru', processors='tokenize,ner')
  doc = nlp(text)
  print(*[f'entity: {ent.text}\ttype: {ent.type}' for sent in doc.sentences for ent in sent.ents], sep='\n')
```

# Почему stanza
Я решил взять эту библиотеку потому что она достаточно новая, и я про нее раньше не слышал и решил попробовать.
А также она действительно многоязычна (заявлено 66 штук в документации)и умеет даже размечать иностранные слова в русском тексте!
Universitas Petropolitana	type: LOC
# Почему такой кусок
Решил взять что то далекое, и как мне кажется в дореволюционные времена люди писали дневники более серьезно и полно, с бОльшим количеством размышлений, в общем, хотелось полноты 
# Какие ошибки
В основном ошибки связаны со словами с большой буквы - часто, если например есть слово Миф - его расценят как персону (у всех ли современных NER теггеров есть проблема с чувствительностью к регистру?) Также я заметил, что доставляет неудобства нестандартная пунктуация, наверное ее стоит попробовать убрать.
Думаю у всех современных NER теггеров есть проблема с чувствительностью к регистру.
Хотелось бы еще найти омонимичные слова, такие как Вашингтон - непонятно, город это или фамилия, и посмотреть, как сработает модель, но таких не нашлось в дневниках.
# Оценка 
Для оценки качества в таких задачах насколько я знаю используется f-мера (среднее гармоническое точности и полноты), но пока нет вручную размеченного золотого стандарта (но его можно в прицнипе сделать и посмотреть что будет, и думаю для первого анализа даже обычная accuracy сойдет)
# Что я думаю
На первый взгляд такое ощущение, что не очень сильно ломатеся современный NER со старым языком, такие же проблемы могли бы возникнуть и с совремнным текстом. Есть очень странные сочетания записанные в тег MISC, помимо производных от названия национальностей, туда попадают совершенно рандомные словосочетания:
